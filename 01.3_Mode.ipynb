# Mode

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

ds1 = pd.read_csv(r"H:\Programming\Files\Statistics\titanic.csv") # ds ==> Dataset
ds = ds1.ffill(axis=0)
ds
```

# Mode

```python
ds["Age"].mode()
```

```python
ds["Age"].mode()[0]
```

```python
ds['Age'].value_counts()
```

```python
mod = ds["Age"].mode()[0]  # mod ==> mode

# Plot the histogram of the "Age" column with bins of width 10
sns.histplot(x="Age", data=ds, bins=[i for i in range(0, 90, 10)])


# [mn] * 300 creates a list of the same x-value (mn) to form a vertical line
# range(300) provides the corresponding y-values
# Plot a vertical red line at the mode age, with label  ==> c="green"
plt.plot([mod] * 300, list(range(300)), c="green", label=f"Mod Age = {mod:.2f}")

# Add legend to show the label
plt.legend()

# Show the plot
plt.show()
```

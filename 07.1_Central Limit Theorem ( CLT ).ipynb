# Central Limit Theorem ( CLT )

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
pop_data = [np.random.randint(10,100) for i in range(10000)]
pop_data
```

```python
pop_table = pd.DataFrame({"pop_data":pop_data})
pop_table
```

# Graph

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="pop_data", data=pop_table)
plt.show()
```

# Making a sample dataset

```python
sample_mean = []

for i in range(50):
    sample_data = []
    for data in range(1000):
        sample_data.append(np.random.choice(pop_data))

    sample_mean.append(np.mean(sample_data))

sample_mean

    
```

```python
sample_M = pd.DataFrame({"sample_mean":sample_mean})

sample_M
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="sample_mean",data=sample_M)
plt.show()
```

# Mean Check

```python
# population data mean
pop = np.mean(pop_data)
# sample data maen 
sam_M = np.mean(sample_data)

pop, sam_M  # allmost same....
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

ds1 = pd.read_csv(r"H:\Programming\Files\Statistics\titanic.csv") # ds ==> Dataset
ds = ds1.ffill(axis=0)
ds
```

# Mean,Median,Mode Graph

```python
mn = ds["Age"].mean()  # mn ==> mean
mdn = ds["Age"].median()  # mdn ==> mean
mod = ds["Age"].mode()[0]  # mod ==> mode

sns.histplot(x="Age", data=ds, bins=[i for i in range(0, 90, 10)])

plt.plot([mn] * 300, list(range(300)), c="red", label=f"Mean Age = {mn:.2f}")
plt.plot([mdn] * 300, list(range(300)), c="blue", label=f"Median Age = {mdn:.2f}")
plt.plot([mod] * 300, list(range(300)), c="green", label=f"Mod Age = {mod:.2f}")

plt.legend()
plt.show()
```

# Median

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds1 = pd.read_csv(r"H:\Programming\Files\Statistics\titanic.csv") # ds ==> Dataset
ds1
```

# Data Cleaning

### Check NaN 

```python
ds1.isnull().sum()
```

```python
# Forward fill missing values along rows (axis=0)
ds = ds1.ffill(axis=0)
ds
```

```python
ds.isnull().sum()
```

# Median

```python
ds["Age"].median()
```

```python
np.median(ds["Age"])
```

```python
sns.histplot(x="Age", data=ds)
sns.set_style("darkgrid")
plt.show()
```

# bins=

```python
sns.histplot(x="Age", data=ds, 
             bins=[i for i in range(0,90,10)])
plt.show()
```

# Calculate the Median age

```python
mdn = np.median(ds["Age"])  # mdn ==> mean

# Plot the histogram of the "Age" column with bins of width 10
sns.histplot(x="Age", data=ds, bins=[i for i in range(0, 90, 10)])


# [mn] * 300 creates a list of the same x-value (mn) to form a vertical line
# range(300) provides the corresponding y-values
# Plot a vertical red line at the mode age, with label  ==> c="blue"
plt.plot([mdn] * 300, list(range(300)), c="blue", label=f"Median Age = {mdn:.2f}")

# Add legend to show the label
plt.legend()

# Show the plot
plt.show()
```

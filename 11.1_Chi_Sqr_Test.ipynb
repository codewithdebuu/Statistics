# Chi Sqr Test

```python
import numpy as np
import scipy.stats as st
```

```python
from IPython.display import Image, display
display(Image(filename=r"H:\Programming\Files\Statistics\pic\Chi_Sqr_Test_01.png"))
```

# Data

```python
Ob = np.array([22,17,20,26,22,13])
Ex = np.array([20,20,20,20,20,20,])
alpha = 0.05
df = (6-1)  # ==> Degrees of freedom(df) =  (n - 1)
```

# Chi Table

```python
chi_Table = st.chi2.ppf((1 - alpha), df)
chi_Table
```

# Chi Sqr Test

```python
chi_cal = np.sum((np.square(Ob - Ex)) / Ex )
chi_cal
```

# Decision

```python
print("Ha is right." if (chi_cal > chi_Table) else "Ho is right.")

# Ho = Die is fair. 
# Ha = Die is not fair.
```

# Data

```python
row_1 = np.array([40,45,25,10])
row_2 = np.array([35,30,20,30])
num_of_sam = 235
alpha = 0.05
```

### Degrees of freedom(df)
df = (row - 1) * (col - 1)

```python
df = (2 - 1) * (4 - 1)
df
```

# Chi Table

```python
chi_Table = st.chi2.ppf((1 - alpha), df)
chi_Table
```

# Chi Value Calculate

```python
sum_row_1 = np.sum(row_1)
sum_row_2 = np.sum(row_2)

sum_row_arr = np.array([sum_row_1, sum_row_2])

sum_row_arr
```

```python
sum_cal = row_1 + row_2
sum_cal
```

## Ex value

```python
Ex = []
for i in sum_row_arr: 
    for j in sum_cal:    
        value = (i * j) / num_of_sam
        Ex.append(value)

Ex
```

### Ob value

```python
Ob = np.array([40,45,25,10,35,30,20,30])
Ob
```

# Chi Sqr Test

```python
chi_cal = np.sum((np.square(Ob - Ex)) / Ex )
chi_cal
```

# Decision

```python
print("Ha is right." if (chi_cal > chi_Table) else "Ho is right.")

# Ho = No association.
# Ha = Association.
```

# Mean absolute Deviation (MAD)

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr_a = np.array([75,65,73,68,72,67])
arr_b = np.array([90,47,43,96,93,51])
arr_c = np.array([1,2,3,4,5,6])
```

```python
mean_a = np.mean(arr_a)
mean_b = np.mean(arr_b)

mean_a, mean_b  # Here mean are same...
```

# mean Graph 

```python
plt.scatter(arr_a, arr_c, label="A")
plt.legend()  # Optional, but needed to display the label
plt.show()
```

# Data Spread Around Their Mean

```python
arr_d = np.array([70,70,70,70,70,70])
mean_d = np.mean(arr_d)

plt.scatter(arr_a, arr_c, label="arr_a")
plt.scatter(arr_b, arr_c, label="arr_b")
plt.plot(arr_d,arr_c, color='gray', linestyle='--', label=f"Mean of arr_d = {mean_d:.2f}")

plt.legend()
plt.show()
```

# Shortcuts

```python
# mean_a & mean_b same ( 70 ). so that we can say,,
mean = mean_a = mean_b

mad_a = (np.sum(np.abs(arr_a - mean)))/ len(arr_a)
mad_b = (np.sum(np.abs(arr_b - mean)))/ len(arr_b)

mad_a, mad_b # in output - Smaller value is batter to use.
# (3.3333333333333335, 23.0) here mad_a is smaller . So that aarr_a is near the mean.
```

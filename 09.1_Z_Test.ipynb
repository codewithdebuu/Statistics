# Z Test

```python
import scipy.stats as st 
import numpy as np
```

```python
from IPython.display import  Image,display
display(Image(filename=r"H:\Programming\Files\Statistics\pic\Z_score_01.png"))
```

# Data

```python
sam_x = 90 # pop => sample mean # [ sample_mean > population_mean ] ==> So that we use Right Tailed.
pop_u = 82 # population mean clamed (Nue)
pop_stu = 20 # population starnderd devision(sigma)
num_of_sam = 81    # [ n > 30 ] ==> So we used Z-Test
alpha = 0.05 

```

# Z_Table 
## st.norm.ppf (1 - alpha)
Z-Table value by alpha value

```python
z_table = st.norm.ppf(1 - alpha)
z_table
```

# Z_Test

```python
z_cal = (sam_x - pop_u)/(pop_stu / np.sqrt(num_of_sam))

z_cal
```

```python
print(f"Z-Table: {z_table} \nZ-cal: {z_cal}")
```

# Decision

```python
print("Ha is right." if (z_table < z_cal) else "Ho is right.")

# Ho ==> Previous
# Ha ==> After
```

```python
old_design_data = np.random.randint(30,55,30)
new_design_data = np.random.randint(40,60,30)

print(old_design_data,"\n")
print(new_design_data)
```

```python
mean_old = np.mean(old_design_data)
mean_new = np.mean(new_design_data)

mean_old,mean_new
```

# Data

```python
pop_stu = 2.5 
num_of_sam = len(new_design_data)
alpha = 0.05 
```

# Z_Table

```python
z_table = st.norm.ppf(1-alpha)
z_table
```

# Z_Test

```python
z_cal = (mean_new - mean_old) / (pop_stu / np.sqrt(num_of_sam))
z_cal
```

# Decision

```python
if (z_cal > z_table):
    print("Ha is right.\nThe new website is batter than old website.")
else:
    print("Ho is right.\nThe new website is not batter than old website.")
```
